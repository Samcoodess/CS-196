Today, we will learn inheritance with this assignment.

```python
class Younger:
  ''' A class Younger with one method greet'''
  def greet(self):
    '''A method to greet '''
    print("Dap me up!")

class Older:
  ''' A class Older with one method greet '''
  def greet(self):
    ''' A method to greet '''
    print("Hello :)")

class PS:
  ''' A class representing Psychology department '''
  dept= "Psychology"
  pass

class CS:
  '''A class representing Computer Science department '''
  dept= "Computer Science"
  pass

class Person:
  """ A class representing person

        Parameters:
        name (str): The name of the person.
        """
  def __init__(self,name):
    self.name = name

class Student(Person,Younger):
  ''' A child class Student inheritated from Parents class Person and Younger '''
  def greet(self):
    '''A method to greet a student.'''
    print(f"You greet {self.name}")
    super().greet() #super() builtin allows methods of base class to be called

class Instructor(Person,Older):
  ''' A child class Instructor inheritated from Parents class Person and Older '''
  def greet(self):
    ''' A method to greet Instructor '''
    print(f"You greet {self.name}")
    super().greet()


class CSStudent(Student,CS):
  '''A class representing a Computer Science student.'''
pass

class PSStudent(Student,PS):
  ''' A class representing a Psychology student. '''
  pass

class CSInstructor(Instructor,CS):
  ''' A class representing a Computer Science Instructor '''
  pass

class PSInstructor(Instructor,PS):
  ''' A class representing a Psychology Instructor. '''
  pass
```

```python
# Create instances of the CSStudent and PSStudent classes
student1=CSStudent('Lilian')
student2=PSStudent('Kail')

# Create instances of the CSInstructor and PSInstructor classes
prof1=CSInstructor('V') 
prof2=PSInstructor('Tsoi')
```

```python
# Call the greet method on the student1 object
student1.greet()
```

```python
# Print the name and department of the student1 object
print(student1.name,'is in',student1.dept)
```

```python
# Call the greet method on the student2 object
student2.greet() 

# Print the name and department of the student2 object
print(student2.name,'is in',student2.dept)
```

```python
# Call the greet method on the prof1 object
prof1.greet()

# Print the name and department of the prof1 object
print('Dr.',prof1.name,'is in',prof1.dept) 
```

```python
# Call the greet method on the prof2 object
prof2.greet() 

# Print the name and department of the prof2 object
print('Dr.',prof2.name,'is in',prof2.dept)
```
