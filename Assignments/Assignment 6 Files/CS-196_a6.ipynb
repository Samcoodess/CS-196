```python
#Open a file myfile.txt ; it's in txt format
f = open('myfile.txt','w')

#write few lines in myfile.txt
f.write('Welcome to your medication tracker!!\n')
f.write('Here you will be updating your medication\n')

f.flush
f.close()
```

```python
#Define a list of data
mydata = ['Presciptions', 'Lab reports', 'Medication']

#Open the file in append(a) mode
f = open('myfile.txt','a')  

#Writelines will write each element and add a newline at the end
f.writelines( txt+'\n' for txt in mydata )

#Flush ensures all data is written to disk, we don't loose data
f.flush
f.close()
```

```python
from dataclasses import dataclass

#Define a custom class using a dataclass decorator
@dataclass
class Prescriptions:
    """A class representing a prescription, with variables for name, dose_mg, and time."""
    name: str
    dose_mg: int
    time: str

#create a object under the prescriptions class
data = {
    'Patient 1': Prescriptions('Antibiotics',20 ,'At Morning before breakfast'),
    'Patient 2': Prescriptions('Bob', 5,'At Night after dinner'),
    'Patient 3': Prescriptions('Elena', 10, 'Two times a day')
}

```

```python
#Import the pickle library
import pickle 

# Create a dictionary
patient = {
    'Patient 1': {
        'Name': "Bob", 'Age' :20, 'Gender':'M'
    },
    'Patient 2': {
        'Name':'Rob', 'Age':21, 'Gender':'M'
    },
    'Patient 3': {
        'Name':'Elena', 'Age':17, 'Gender':'F'
    }
}
#Use pickle to dump the data into a binary file called 'patients.pkl'
with open('patients.pkl','wb') as f: #open the file in wb(binary write) mode)
    pickle.dump(data, f)
```

```python
#open the file in rb(read binary)mode
with open('patients.pkl','rb') as f:
    data = pickle.load(f)
```

```python
# Import the `json` library to work with JSON data
import json

# Create a dictionary of patient data
patient_data = {
    'Patient 1': {
        'name': 'Antibiotics',
        'dose_mg': 20,
        'time': 'At Morning before breakfast'
    },
    'Patient 2': {
        'name': 'Bob',
        'dose_mg': 5,
        'time': 'At Night after dinner'
    },
    'Patient 3': {
        'name': 'Elena',
        'dose_mg': 10,
        'time': 'Two times a day'
    }
}          
```

```python
type(patient_data)
```

```python
#use json library to dump the data into a json file
with open('patients.json','w') as f:
    json.dump(patient_data, f)

with open('patients.json', 'r') as f:
    data2 = json.load(f)
```

```python
print(data2)
```

```python
#printing data2 in nicer way
for key, value in data2.items():
    print(key + ':', end=' ')
    for k, v in value.items():
        print(k + ':', v, end=' ')
    print()
```
