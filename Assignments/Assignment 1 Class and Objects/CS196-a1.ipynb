```python
#Define a class Dog
class Dog:
    '''Dog class has name,age,breed.'''

    def __init__(self, name,age,breed):
        self.name = name
        self.age = age
        self.breed = breed

    def talk(self):
      '''Prints the name and the action for talk.'''
      print(f'{self.name} says woof')

    def come(self):
      '''Print the name and the action for come.'''
      print(f'{self.name} runs over')

    def sit(self):
      '''Print the name and the action for sit.'''
      print(f'{self.name} sits')

```

```python
#@title Default title text
# create an objects called dog1 and dog2, which are instances of class Dog

dog1 = Dog('Fifi',4,'Samoyeed')
dog2 = Dog('Kiki',6,'Bulldog')
dog3 = Dog()
```

Calling the function for two objects of class Dog

```python

dog1.talk()
dog2.talk()
```

```python
dog1.come()
dog2.come()
```

```python
dog1.sit()
dog2.sit()
```

```python
print(f"The {dog1.breed} named {dog1.name} is {dog1.age} years old now.")
```

#For Cat

```python
class Cat:
    '''Dog class has name,age,breed.'''

    def __init__(self, name,age,breed):
        self.name = name
        self.age = age
        self.breed = breed

    def talk(self):
      '''Prints the name and the action for talk.'''
      print(f'{self.name} says meow')

    def come(self):
      '''Prints the name and the action for come.'''
      print(f'{self.name} lazily meanders over')

    def sit(self):
      '''Prints the name and the action for sit.'''
      print(f'{self.name} gives you the a contemptous look')
```

```python
# create an objects called cat1 and cat2, which are instances of class Cat

cat1= Cat("Cherry",2,"briman")
cat2 = Cat("Mimi",1,"toyger")
```

Calling the function for two objects of class Cat

```python
cat1.talk()
cat2.talk()
```

```python
cat1.come()
cat2.come()
```

```python
cat1.sit()
cat2.sit()
```
