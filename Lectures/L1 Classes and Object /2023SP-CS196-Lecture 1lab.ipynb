##### sets

```python
s = {"alex", "ben", "claire"}
s.add("alex")
s.add("claire")
s.remove('ben')
print( len(s) )
```

```python
s1 = {1,2,3,4}
s2 = {3,4,5,6}
s3 = s1.union(s2)
s4 = s1.intersection(s2)
print( s3 )
print( s4 )
```
